# Reading from file

```python
import pandas as pd

df = pd.read_csv("./admission_algorithms_dataset.csv")

#print(df)
```

```python
df
```

```python
# vectorized operations
norm_sat = (df["SAT"] / 1600 )
norm_gpa = (df["GPA"] / 5)
avg_sat_gpa = (norm_sat + norm_gpa) / 2
avg_sat_gpa
df["AVG_SAT_GPA"] = avg_sat_gpa
```

```python
df
```

# Using boolean operations on columns

```python
# Grab row index 0
row = df.loc[0]
is_top_gpa = row["GPA"] >= 4.0
is_top_sat = row["SAT"] >= 1300.0

# on individuals row, can use "or", "and", "not" for boolean expressions
is_top_both = (is_top_sat and is_top_gpa)
is_top_both

# on column I must bitwise operations that "|" (for or)
is_top_gpa_all = df["GPA"] >= 4.0
is_top_sat_all = df["SAT"] >= 1300.0
is_top_both_all = (is_top_gpa_all & is_top_sat_all)
is_top_both_all
```
